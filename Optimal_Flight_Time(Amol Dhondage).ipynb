Handling Outliers

```python
plt.figure(figsize=(30,15))
sns.boxplot(data['Price'],color='Red')
plt.title('Checking for Outliers', fontsize = 30)
plt.xlabel('Price', fontsize = 20)
plt.show()
```

**There are more number of outliers present in the price column**

**Now removing the outliers using Inter Quartile range method**

```python
Q1 = np.quantile(data['Price'],0.25)
Q3 = np.quantile(data['Price'],0.75)
```

```python
IQR = Q3-Q1
IQR
```

```python
lower_limit = Q1-1.5*IQR
upper_limit = Q3 + 1.5*IQR
```

```python
data_cleaned = data[(data['Price']<np.quantile(data['Price'],0.795))&(data['Price']>lower_limit)]
```

```python
data_cleaned
```

```python
plt.figure(figsize=(30,15))
sns.boxplot(data_cleaned['Price'],color='Red')
plt.title('Checking for Outliers', fontsize = 30)
plt.xlabel('Price', fontsize = 20)
plt.show()
```
